## Checking out the outliers and removing them

```python
sns.boxplot(df['price'])
```

```python
median_price = df['price'].median()
df["price"] = np.where(df["price"] > 12000000, median_price, df['price']) 
sns.boxplot(df['price'])
```

```python
sns.boxplot(df['area'])
```

```python
median_area = df['area'].median()
df["area"] = np.where(df["area"] > 14000, median_area, df['area']) 
sns.boxplot(df['area'])
```

## Checking for categorical columns and perform Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['furnishingstatus'] = le.fit_transform(df['furnishingstatus'])
df['mainroad'] = le.fit_transform(df['mainroad'])
df['guestroom'] = le.fit_transform(df['guestroom'])
df['basement'] = le.fit_transform(df['basement'])
df['hotwaterheating'] = le.fit_transform(df['hotwaterheating'])
df['airconditioning'] = le.fit_transform(df['airconditioning'])
```

```python
df.head()
```

```python
X= df.iloc[:,1:]
X
```

```python
Y = df.iloc[:,0]
Y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_std = scaler.fit_transform(X)
```

```python
x_std
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state =2)
```

```python
print(x_train.shape)
x_train.head()
```

```python
print(x_test.shape)
x_test.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr = LinearRegression()
```

```python
lr.fit( x_train, y_train )
```

```python
print("Value of the coefficients: \n", lr.coef_)
print("Value of the intercept: \n", lr.intercept_)
```

```python
Y_pred = lr.predict(x_test)
```

```python
print("MSE",mean_squared_error(y_test,Y_pred))
print("RMSE",np.sqrt(mean_squared_error(y_test,Y_pred)))
print("R-Square", r2_score(y_test,Y_pred))
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
pred1=lr.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(y_test,pred))
confusion_matrix(y_test,pred)
```

```python
print(accuracy_score(y_train,pred1))
print(accuracy_score(y_test,pred))
```

```python
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
pred1=lr.predict(x_train)
```

```python
print(accuracy_score(y_train,pred1))
print(accuracy_score(y_test,pred))
```

```python
#SVM
from sklearn.svm import SVC
model1=SVC(kernel="rbf")
msvc=model1.fit(x_train,y_train)
pred=msvc.predict(x_test)
```

```python
print(confusion_matrix(y_test,pred))
print(accuracy_score(y_test,pred))
```
